```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# the invenroty file gives a glance of all '.txt' files
df_inventory = pd.read_csv("encyclopaediaBritannica-inventory.csv", header=None)
df_inventory.columns = ['file_name','information']
df_inventory
```

```python
# how many words in one .txt file? (usually one txt file represents one volume of Encyclopaedia Britannica)

list = []
# Traverse the files to count the words_num in each .txt file
for index, row in df_inventory.iterrows():
    f = open('txt/' + row['file_name'], 'r', encoding="utf8")
    c = f.read()
    num = c.split()
    list.append(len(num)) # the number of words for each .txt file is stored in 'list'
    f.close()
    
df_inventory = df_inventory.assign(words_num = list)
df_inventory
```

```python
print('in one volume of Encyclopaedia Britannica')
print('number of the most words is: ', df_inventory['words_num'].max())
print('number of the least words is: ', df_inventory['words_num'].min())
print('number of the average words is: ', round(df_inventory['words_num'].mean(),2))
```

```python
# distribution of words_num across all volumes of eight editions of Encyclopaedia Britannica

plt.figure(figsize=(10, 5))
chart = sns.distplot(df_inventory['words_num'])
chart.set_title('Distribution of words_num across all volumes')
plt.show()
```
